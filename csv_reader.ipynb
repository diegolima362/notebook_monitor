```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import pandas as pd
from pandas.plotting import autocorrelation_plot
```

```python
csv_data = pd.read_csv("simulation01.csv")

del csv_data['_']
del csv_data['__']
del csv_data['___']
del csv_data['instanceId']
del csv_data['value_decimal']

csv_data.time = pd.to_datetime(csv_data.time)

etco2 = csv_data[csv_data.metricID != 'MDC_PULS_OXIM_SAT_O2']
spo2 = csv_data[csv_data.metricID == 'MDC_PULS_OXIM_SAT_O2']

etco2 = etco2.rename(columns={"value": "etco2"})
spo2 = spo2.rename(columns={"value": "spo2"})
```

```python
figure(figsize=(20, 10), dpi=192)
fig = spo2.plot(x ='time', y='spo2', kind = 'line').get_figure()
plt.axhline(94, color="red", linestyle="-.", label="Negligible")
plt.axhline(91, color="red", linestyle="--", label="Minor")
plt.axhline(85, color="red", linestyle=":", label="Serious")
plt.legend(loc=0)
```
